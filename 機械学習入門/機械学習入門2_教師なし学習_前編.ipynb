```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#日本プロ野球の2019年の投手の防御率ランキング(規定投球回数が1/2以上の選手を出力)
#こちらのデータをもとにcsvを作成 → https://baseball-data.com/19/stats/pitcher-all/era-3.html
npb= pd.read_csv('ML_data/投手成績2019.csv')
#上位50人を出力
npb.head(20)
```

```python
npb.shape
```

```python
npb.mean()
```

```python
npb.loc[:,"防御率"].mean()
```

```python
plot_npb = npb.rename(columns={'防御率': 'DefensiveRate',
                             '勝率': 'WinRate',
            })
plot_npb
```

```python
sns.pairplot(plot_npb[["DefensiveRate", "WinRate", "DIPS"]])
plt.show()
```

```python
correlation = plot_npb[["DefensiveRate", "WinRate", "DIPS"]].corr()
sns.heatmap(correlation, annot=True)
```
