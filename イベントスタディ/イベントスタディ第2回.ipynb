<h3>データを読み込む</h3>

```python
import numpy as np
import pandas as pd
```

```python
#キリンホールディングスの自社株買い前後の累積異常リターン（CAR）を求める
df = pd.read_csv("stock_data/2503_2019.csv", encoding="shift-jis") 
df
```

<h3>マーキングをする</h3>

```python
#キリンの自社株買いの実行日
df[df["日付"] == "2019/11/8"] 
```

```python
#自社株買い実行日をt＝０としたときの他の日のt
df["t"]= [i for i in range(-204, 37)]
df
```

```python
df[df["日付"] == "2019/11/8"] 
```

<h3>リターンを求める</h3>

```python
df["r_i"] = (df["終値調整値"] - df["終値調整値"].shift(1))/ df["終値調整値"].shift(1)
df
```

```python
#1306 東証ETF TOPIX連動型上場投資信託（ETF）
etf = pd.read_csv("stock_data/1306_2019.csv", encoding="shift-jis") 
etf
```

```python
etf["r_m"] = (etf["終値調整値"] - etf["終値調整値"].shift(1))/ etf["終値調整値"].shift(1)
etf
```

```python
df["r_m"] = etf["r_m"]
df
```

<h3>回帰分析をおこないマーケットモデルを作成する</h3>

```python
#推定期間スタート
df[df["t"] == -200]
```

```python
#推定期間エンド
df[df["t"] == -21]
```

```python
kirin = df[["r_i"]][4:183]
kirin
```

```python
market = df[["r_m"]][4:183]
market
```

```python
from sklearn import linear_model # 線形モデル関係
import statsmodels.api as smf # 統計量計算

Y = kirin["r_i"]
X = market["r_m"]

#単回帰分析モデルの作成
model = smf.OLS(Y,X) 
result = model.fit()
result.summary()
```

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()

Y = kirin[["r_i"]]
X = market[["r_m"]]

clf.fit(X,Y)

print(f"傾き : {clf.coef_[0][0]}")

print(f"切片 : {clf.intercept_[0]}")

#回帰式
print(f"回帰式 : y = {clf.intercept_[0]} + {clf.coef_[0][0]}x")
```

<h3>ベンチマークを計算する</h3>

```python
df[df["t"] == -20]
```

```python
df["expect"] = list(map(lambda x: clf.intercept_[0] + clf.coef_[0][0] * x, df["r_m"].values))
df
```
