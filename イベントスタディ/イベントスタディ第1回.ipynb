```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#1306 東証ETF TOPIX連動型上場投資信託（ETF）
df =pd.read_csv("stock_data/1306_2019.csv", encoding="shift-jis")
df.set_index("日付")
```

```python
plt.figure(figsize=(12.2,4.5)) #width = 12.2in, height = 4.5
plt.plot(df.index,df["終値調整値"],  label='Close Price')
plt.title('Close Price History')
plt.xlabel('January 4, 2019 - December 30, 2019',fontsize=18)
plt.ylabel('Close. Price Yen (\)',fontsize=18)
plt.show()
```

```python
df["return"] = (df["終値調整値"] - df["終値調整値"].shift(1))/ df["終値調整値"].shift(1)
```
